### Goals Scored & Concede at the end of Matchweek, Arranged by Teams & Matchweek

```python
# Fungsi untuk mendapatkan jumlah gol yang dicetak (goals scored)

# oleh setiap tim di setiap pekan pertandingan
def get_goals_scored(playing_stat):

    # Inisialisasi dictionary dengan nama tim unik dari kolom HomeTeam dan AwayTeam
    teams = {team: [] for team in pd.concat([playing_stat['HomeTeam'],
                                             playing_stat['AwayTeam']]).unique()}
    # Loop untuk mengisi jumlah gol yang dicetak oleh tim tuan rumah dan tim tamu
    for i in range(len(playing_stat)):
        HTGS = playing_stat.iloc[i]['FTHG']  # Gol yang dicetak oleh tim tuan rumah
        ATGS = playing_stat.iloc[i]['FTAG']  # Gol yang dicetak oleh tim tamu
        teams[playing_stat.iloc[i].HomeTeam].append(HTGS)
        teams[playing_stat.iloc[i].AwayTeam].append(ATGS)

    # Buat DataFrame untuk menyimpan jumlah gol yang dicetak per pekan
    GoalsScored = pd.DataFrame(data=teams, index=[i for i in range(1, 39)]).T
    GoalsScored[0] = 0  # Tambahkan kolom awal untuk agregasi kumulatif

    # Hitung jumlah gol yang dicetak secara kumulatif hingga pekan tertentu
    for i in range(2, 39):
        GoalsScored[i] = GoalsScored[i] + GoalsScored[i - 1]

    return GoalsScored


# Fungsi untuk mendapatkan jumlah gol yang kebobolan (goals conceded)
# oleh setiap tim di setiap pekan pertandingan
def get_goals_conceded(playing_stat):
    # Inisialisasi dictionary dengan nama tim unik dari kolom HomeTeam dan AwayTeam
    teams = {team: [] for team in pd.concat([playing_stat['HomeTeam'], playing_stat['AwayTeam']]).unique()}

    # Loop untuk mengisi jumlah gol yang kebobolan oleh tim tuan rumah dan tim tamu
    for i in range(len(playing_stat)):
        HTGC = playing_stat.iloc[i]['FTAG']  # Gol yang kebobolan oleh tim tuan rumah
        ATGC = playing_stat.iloc[i]['FTHG']  # Gol yang kebobolan oleh tim tamu
        teams[playing_stat.iloc[i].HomeTeam].append(HTGC)
        teams[playing_stat.iloc[i].AwayTeam].append(ATGC)

    # Buat DataFrame untuk menyimpan jumlah gol yang kebobolan per pekan
    GoalsConceded = pd.DataFrame(data=teams, index=[i for i in range(1, 39)]).T
    GoalsConceded[0] = 0  # Tambahkan kolom awal untuk agregasi kumulatif

    # Hitung jumlah gol yang kebobolan secara kumulatif hingga pekan tertentu
    for i in range(2, 39):
        GoalsConceded[i] = GoalsConceded[i] + GoalsConceded[i - 1]

    return GoalsConceded

# Fungsi untuk menambahkan kolom data gol yang dicetak dan kebobolan
# ke dalam DataFrame utama `playing_stat`
def get_gss(playing_stat):
    # Buat salinan DataFrame untuk menghindari perubahan pada data asli
    playing_stat = playing_stat.copy()

    # Hitung gol yang kebobolan dan dicetak menggunakan fungsi sebelumnya
    GC = get_goals_conceded(playing_stat)
    GS = get_goals_scored(playing_stat)

    # Variabel untuk mencatat jumlah pekan pertandingan
    j = 0
    HTGS = []  # Home Team Goals Scored
    ATGS = []  # Away Team Goals Scored
    HTGC = []  # Home Team Goals Conceded
    ATGC = []  # Away Team Goals Conceded

    # Loop untuk mengisi data gol berdasarkan tim tuan rumah dan tim tamu
    for i in range(380):  # Total pertandingan dalam satu musim adalah 380
        ht = playing_stat.iloc[i].HomeTeam
        at = playing_stat.iloc[i].AwayTeam
        HTGS.append(GS.loc[ht][j])
        ATGS.append(GS.loc[at][j])
        HTGC.append(GC.loc[ht][j])
        ATGC.append(GC.loc[at][j])

        # Perbarui indeks pekan setiap 10 pertandingan
        if ((i + 1) % 10) == 0:
            j = j + 1

    # Tambahkan kolom hasil perhitungan ke DataFrame
    playing_stat = playing_stat.assign(
        HTGS=HTGS,
        ATGS=ATGS,
        HTGC=HTGC,
        ATGC=ATGC
    )

    return playing_stat
```

### Get Respective Points

```python
# Fungsi untuk mengonversi hasil pertandingan menjadi poin
def get_points(result):
    """
    Mengembalikan jumlah poin berdasarkan hasil pertandingan.
    - 'W' (Win): 3 poin
    - 'D' (Draw): 1 poin
    - 'L' (Loss): 0 poin
    """
    if result == 'W':
        return 3
    elif result == 'D':
        return 1
    else:
        return 0


def get_cuml_points(matchres):
    """
    Menghitung poin kumulatif berdasarkan hasil pertandingan setiap tim.
    """
    # Ubah hasil pertandingan menjadi poin menggunakan apply
    matchres_points = matchres.apply(lambda row: row.map(get_points), axis=1)  # Operasi pada tiap baris

    # Hitung poin kumulatif hingga pekan tertentu
    for i in range(2, 39):
        matchres_points[i] = matchres_points[i] + matchres_points[i - 1]

    # Tambahkan kolom awal dengan nilai 0 untuk pekan sebelum liga dimulai
    matchres_points.insert(column=0, loc=0, value=[0 for _ in range(len(matchres_points))])

    return matchres_points


# Fungsi untuk mendapatkan hasil pertandingan (Win, Draw, Loss) setiap tim
def get_matchres(playing_stat):
    """
    Menghasilkan DataFrame hasil pertandingan setiap tim dalam format:
    - 'W' (Win): Menang
    - 'D' (Draw): Seri
    - 'L' (Loss): Kalah
    """
    # Buat dictionary dengan tim sebagai key dan daftar hasil pertandingan sebagai value
    teams = {team: [] for team in pd.concat([playing_stat['HomeTeam'], playing_stat['AwayTeam']]).unique()}

    # Iterasi setiap pertandingan untuk mengisi hasil pertandingan
    for i in range(len(playing_stat)):
        if playing_stat.iloc[i].FTR == 'H':  # Tuan rumah menang
            teams[playing_stat.iloc[i].HomeTeam].append('W')
            teams[playing_stat.iloc[i].AwayTeam].append('L')
        elif playing_stat.iloc[i].FTR == 'A':  # Tim tamu menang
            teams[playing_stat.iloc[i].AwayTeam].append('W')
            teams[playing_stat.iloc[i].HomeTeam].append('L')
        else:  # Pertandingan seri
            teams[playing_stat.iloc[i].AwayTeam].append('D')
            teams[playing_stat.iloc[i].HomeTeam].append('D')

    # Konversi dictionary ke DataFrame dengan tim sebagai baris dan pekan sebagai kolom
    return pd.DataFrame(data=teams, index=[i for i in range(1, 39)]).T


# Fungsi utama untuk menambahkan kolom poin kumulatif ke DataFrame utama
def get_agg_points(playing_stat):
    """
    Menambahkan kolom poin kumulatif untuk tim tuan rumah (HTP) dan tim tamu (ATP)
    ke dalam DataFrame `playing_stat`.
    """
    # Dapatkan hasil pertandingan dan poin kumulatif
    matchres = get_matchres(playing_stat)
    cum_pts = get_cuml_points(matchres)

    # Inisialisasi list untuk menyimpan poin kumulatif
    HTP = []  # Home Team Points
    ATP = []  # Away Team Points

    j = 0  # Pekan pertandingan (dimulai dari 0)
    for i in range(len(playing_stat)):
        # Ambil nama tim tuan rumah dan tamu
        ht = playing_stat.iloc[i].HomeTeam
        at = playing_stat.iloc[i].AwayTeam

        # Tambahkan poin kumulatif untuk tim tuan rumah dan tamu
        HTP.append(cum_pts.loc[ht][j])
        ATP.append(cum_pts.loc[at][j])

        # Perbarui indeks pekan setiap 10 pertandingan
        if ((i + 1) % 10) == 0:
            j += 1

    # Tambahkan kolom poin kumulatif ke DataFrame menggunakan `.assign`
    playing_stat = playing_stat.assign(HTP=HTP, ATP=ATP)

    return playing_stat
```

```python
# Terapkan fungsi ke setiap dataset dari playing_statistics_1 sampai playing_statistics_20
for i in range(1, 19):  # Loop dari 1 hingga 18
    dataset_name = f"playing_statistics_{i}"  # Buat nama dataset secara dinamis
    globals()[dataset_name] = get_agg_points(globals()[dataset_name])
```

### Get Team Form

```python
def get_form(playing_stat, num):
    """
    Membuat DataFrame yang berisi bentuk (form) hasil pertandingan terakhir
    dari setiap tim hingga jumlah tertentu (`num`).
    """
    form = get_matchres(playing_stat)  # Dapatkan hasil pertandingan
    form_final = form.copy()  # Salin DataFrame form untuk diproses

    # Loop untuk menggabungkan hasil pertandingan sebelumnya hingga num pekan
    for i in range(num, 39):
        form_final[i] = ''  # Inisialisasi kolom baru untuk bentuk tim
        j = 0
        while j < num:
            form_final[i] += form[i - j]  # Gabungkan hasil pertandingan sebelumnya
            j += 1

    return form_final


def add_form(playing_stat, num):
    """
    Menambahkan kolom bentuk tim (`HM<num>` untuk HomeTeam dan `AM<num>` untuk AwayTeam)
    berdasarkan hasil pertandingan terakhir (`num` pertandingan).
    """
    form = get_form(playing_stat, num)  # Dapatkan bentuk pertandingan
    h = ['M' for _ in range(num * 10)]  # Placeholder untuk tim tuan rumah (belum tersedia untuk num pekan awal)
    a = ['M' for _ in range(num * 10)]  # Placeholder untuk tim tamu (belum tersedia untuk num pekan awal)

    j = num  # Mulai iterasi dari pekan ke-num
    for i in range(num * 10, 380):  # Loop untuk semua pertandingan mulai dari pekan ke-num
        ht = playing_stat.iloc[i].HomeTeam  # Ambil nama tim tuan rumah
        at = playing_stat.iloc[i].AwayTeam  # Ambil nama tim tamu

        # Ambil bentuk pertandingan terakhir tim tuan rumah
        past = form.loc[ht][j]
        h.append(past[num - 1])  # Ambil pertandingan terakhir (0 adalah yang terbaru)

        # Ambil bentuk pertandingan terakhir tim tamu
        past = form.loc[at][j]
        a.append(past[num - 1])  # Ambil pertandingan terakhir (0 adalah yang terbaru)

        # Perbarui indeks pekan setiap 10 pertandingan
        if ((i + 1) % 10) == 0:
            j += 1

    # Tambahkan kolom ke DataFrame utama
    playing_stat = playing_stat.copy()  # Pastikan bekerja pada salinan eksplisit
    playing_stat['HM' + str(num)] = h
    playing_stat['AM' + str(num)] = a

    return playing_stat


def add_form_df(playing_statistics):
    """
    Menambahkan bentuk tim (form) untuk jumlah pertandingan terakhir (1-5)
    ke DataFrame utama.
    """
    playing_statistics = add_form(playing_statistics, 1)
    playing_statistics = add_form(playing_statistics, 2)
    playing_statistics = add_form(playing_statistics, 3)
    playing_statistics = add_form(playing_statistics, 4)
    playing_statistics = add_form(playing_statistics, 5)
    return playing_statistics
```

```python
# Terapkan fungsi ke setiap dataset dari playing_statistics_1 sampai playing_statistics_20
for i in range(1, 19):  # Loop dari 1 hingga 18
    dataset_name = f"playing_statistics_{i}"  # Buat nama dataset secara dinamis
    globals()[dataset_name] = add_form_df(globals()[dataset_name])
```

### Get MatchWeek

```python
def get_mw(playing_stat):
    """
    Menambahkan kolom `MW` (MatchWeek) ke DataFrame `playing_stat`.
    Kolom `MW` merepresentasikan pekan pertandingan, dengan nilai
    yang meningkat setiap 10 pertandingan.
    """
    j = 1  # Awal MatchWeek
    MatchWeek = []  # List untuk menyimpan nilai MatchWeek

    # Loop melalui setiap pertandingan dalam dataset
    for i in range(380):  # Asumsikan ada 380 pertandingan total
        MatchWeek.append(j)  # Tambahkan nilai MatchWeek saat ini
        if ((i + 1) % 10) == 0:  # Tingkatkan pekan setiap 10 pertandingan
            j += 1

    # Pastikan bekerja pada salinan eksplisit DataFrame
    # playing_stat = playing_stat.copy()
    playing_stat['MW'] = MatchWeek  # Tambahkan kolom MatchWeek
    return playing_stat


# Terapkan fungsi ke setiap dataset dari playing_statistics_1 sampai playing_statistics_18
for i in range(1, 19):  # Loop dari 1 hingga 18
    dataset_name = f"playing_statistics_{i}"  # Buat nama dataset secara dinamis
    if dataset_name in globals():  # Periksa apakah dataset ada di namespace global
        globals()[dataset_name] = get_mw(globals()[dataset_name])  # Perbarui dataset dengan MatchWeek
```

### Final Exploration Dataframe

```python
playing_stat = pd.concat([playing_statistics_1,
                          playing_statistics_2,
                          playing_statistics_3,
                          playing_statistics_4,
                          playing_statistics_5,
                          playing_statistics_6,
                          playing_statistics_7,
                          playing_statistics_8,
                          playing_statistics_9,
                          playing_statistics_10,
                          playing_statistics_11,
                          playing_statistics_12,
                          playing_statistics_13,
                          playing_statistics_14,
                          playing_statistics_15,
                          playing_statistics_16,
                          playing_statistics_17,
                          playing_statistics_18,
                          ], ignore_index=True)

# Menghitung total poin berdasarkan string hasil pertandingan sebelumnya (form tim)
def get_form_points(string):
    sum = 0
    for letter in string:
        sum += get_points(letter)  # Menggunakan fungsi get_points untuk setiap huruf dalam string
    return sum

# Menggabungkan hasil pertandingan terakhir (5 pertandingan terakhir) untuk tim tuan rumah dan tim tandang
playing_stat['HTFormPtsStr'] = playing_stat['HM1'] + playing_stat['HM2'] + playing_stat['HM3'] + playing_stat['HM4'] + playing_stat['HM5']
playing_stat['ATFormPtsStr'] = playing_stat['AM1'] + playing_stat['AM2'] + playing_stat['AM3'] + playing_stat['AM4'] + playing_stat['AM5']

# Menghitung total poin dari 5 pertandingan terakhir untuk tim tuan rumah dan tim tandang
playing_stat['HTFormPts'] = playing_stat['HTFormPtsStr'].apply(get_form_points)
playing_stat['ATFormPts'] = playing_stat['ATFormPtsStr'].apply(get_form_points)

# Menampilkan semua nama kolom dalam dataframe playing_stat
playing_stat.keys()

```

```python
# Menghitung selisih gol (goal difference) untuk tim tuan rumah dan tim tandang
# HTGD (Home Team Goal Difference) = Jumlah gol yang dicetak oleh tim tuan rumah - Jumlah gol yang kebobolan oleh tim tuan rumah
playing_stat['HTGD'] = playing_stat['HTGS'] - playing_stat['HTGC']  # HTGS: Home Team Goals Scored, HTGC: Home Team Goals Conceded

# ATGD (Away Team Goal Difference) = Jumlah gol yang dicetak oleh tim tandang - Jumlah gol yang kebobolan oleh tim tandang
playing_stat['ATGD'] = playing_stat['ATGS'] - playing_stat['ATGC']  # ATGS: Away Team Goals Scored, ATGC: Away Team Goals Conceded

# Menghitung selisih poin antara tim tuan rumah dan tim tandang
# DiffPts (Difference in Points) = Poin tim tuan rumah - Poin tim tandang
playing_stat['DiffPts'] = playing_stat['HTP'] - playing_stat['ATP']  # HTP: Home Team Points, ATP: Away Team Points

# Menghitung selisih poin dari performa 5 pertandingan terakhir antara tim tuan rumah dan tim tandang
# DiffFormPts (Difference in Form Points) = Total poin dari form tim tuan rumah - Total poin dari form tim tandang
playing_stat['DiffFormPts'] = playing_stat['HTFormPts'] - playing_stat['ATFormPts']  # HTFormPts: Home Team Form Points, ATFormPts: Away Team Form Points
```

### Melakukan normalisasi kolom terkait dengan Matchweek (MW)

 Tujuannya adalah untuk mengurangi bias terhadap tim yang bermain di lebih banyak pertandingan dengan membagi nilai pada kolom tertentu dengan jumlah Matchweek hingga saat itu.


```python
# Daftar kolom yang akan dinormalisasi
cols = ['HTGD', 'ATGD', 'DiffPts', 'DiffFormPts', 'HTP', 'ATP']

# Pastikan kolom MW (Matchweek) bertipe float untuk mendukung operasi pembagian
playing_stat.MW = playing_stat.MW.astype(float)

# Normalisasi setiap kolom dalam daftar `cols` dengan membagi nilai pada kolom tersebut dengan MW
for col in cols:
    playing_stat[col] = playing_stat[col] / playing_stat.MW  # Nilai dibagi dengan jumlah Matchweek
```

```python
def only_hw(string):
    if string == 'H':
        return 'H'
    else:
        return 'NH'

playing_stat['FTR'] = playing_stat.FTR.apply(only_hw)
```

## Build Model With Pipeline

```python
# Remove few column
dataset3 = df.copy().drop(columns =['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG',
       'HTGS', 'ATGS', 'HTGC', 'ATGC',
       'HM4', 'HM5','AM4', 'AM5', 'MW', 'HTFormPtsStr',
       'ATFormPtsStr', 'HTFormPts', 'ATFormPts',
       'DiffPts'] )
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split

# Memisahkan fitur dan target
X_bestmodel = dataset3.drop('FTR', axis=1)  # Fitur
y_bestmodel = dataset3['FTR']  # Target
```

```python
X_bestmodel.head()
```

```python
# Label Encoding untuk target 'FTR'
target_encoder = LabelEncoder()
y_encoded = target_encoder.fit_transform(y_bestmodel)

# Menyusun preprocessing pipeline
# 1. One-Hot Encoding untuk fitur kategorikal
# 2. Standard Scaler untuk fitur numerik
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['HTGD', 'ATGD', 'HTP', 'ATP', 'DiffFormPts']),  # Fitur numerik
        ('cat', OneHotEncoder(), ['HM1', 'HM2', 'HM3', 'AM1', 'AM2', 'AM3'])       # Fitur kategorikal
    ])
```

```python
from sklearn.linear_model import LogisticRegression

# Membuat pipeline dengan preprocessing dan model (Logistic Regression)
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(
            solver='newton-cholesky',  # Solver yang digunakan
            penalty='l2',              # Regularisasi L2
            C=1.0,                     # Strength of regularization
            max_iter=1000,             # Iterasi maksimum
            random_state=42            # Untuk reproduktifitas
            ))
])
```

```python
# Memecah dataset menjadi train dan test
X_train, X_test, y_train, y_test = train_test_split(X_bestmodel, y_encoded,
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
# Melatih model dengan pipeline
pipeline.fit(X_train, y_train)
```

```python
# Mengevaluasi model pada data test
accuracy = pipeline.score(X_test, y_test)
print(f"Accuracy: {accuracy:.4f}")

# Untuk prediksi dengan data baru
new_data = X_test.iloc[0:1]  # Ambil satu contoh data baru
prediction = pipeline.predict(new_data)
print(f"Prediksi untuk data baru: {prediction}")
```
